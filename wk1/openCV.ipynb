```python
# Import the libraries
import cv2
import matplotlib.pyplot as plt
```

### 2. Upload an image

```python
img_path = './pic/golf.png'
# Read the image
img = cv2.imread(img_path)
```

### 3. Display the image

```python
plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.title('Uploaded Image')
plt.axis('off')
plt.show()
```

### 4. Resize the image

```python
# Convert the image from BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Calculate the original ratio
original_height, original_width = img_rgb.shape[:2]
new_height = 300
aspect_ratio = new_height / original_height
new_width = int(aspect_ratio * original_width)

# Resize the image
img_resized = cv2.resize(img_rgb, (new_width, new_height), interpolation=cv2.INTER_AREA)
```

### 5. Display the resize image

```python
plt.figure(figsize=(10, 10))
plt.imshow(img_resized)
plt.title('Resized Image')
plt.axis('off')
plt.show()
```

### 6. Rotate the image

```python
# Rotate the image 180 degrees
rotated_img = cv2.rotate(img, cv2.ROTATE_180)
```

### 7. Display the rotated image

```python
plt.figure(figsize=(10, 10))
plt.imshow(rotated_img)
plt.title('Rotated Image')
plt.axis('off')
plt.show()
```
